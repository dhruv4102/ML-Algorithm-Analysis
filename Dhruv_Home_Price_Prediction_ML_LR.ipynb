```python
new_df = df.drop(['Address','Date','Postcode','YearBuilt',"Lattitude","Longtitude"],axis='columns')
new_df
```

```python
new_df.isnull().sum()
```

```python
new_data = new_df.fillna({
        'Propertycount': 0,
        'Distance': 0,
        'Bedroom2': 0,
        'Bathroom': 0,
        'Car': 0
    })
new_data
```

```python
new_data['Landsize'] = new_data['Landsize'].fillna(new_data['Landsize'].mean())
new_data['BuildingArea'] = new_data['BuildingArea'].fillna(new_data['BuildingArea'].mean())
new_data
```

```python
new_data.dropna(inplace=True)
```

```python
new_data.Method.unique()
```

```python
new_data.shape
```

```python
final_data = pd.get_dummies(new_data, drop_first=True)
final_data
final_data.shape
```

```python
x = final_data.drop('Price', axis=1)
y = final_data.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=8)
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
```
